```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import math as math



n = 20

X = np.empty(n, dtype=float)
np.random.seed(42)

for i in range(0,n):
    X[i] = np.random.uniform(0,2*math.pi)
X = np.sort(X)
y = np.cos(X) + np.random.randn(n) * 0.3

plt.figure(figsize=(10, 4))

ax = plt.subplot(1, 3, 1)
linear = LinearRegression()
linear.fit(X[:, np.newaxis], y)

X_test = np.linspace(0, 2*math.pi)
plt.plot(X_test, linear.predict(X_test[:, np.newaxis]), label="Vorhersage", color='red')
plt.plot(X_test, np.cos(X_test), label="cos", color='green')
plt.scatter(X, y, edgecolor='b', s=20, label="Datenstze")
plt.xlim((0, 2*math.pi))
plt.ylim((-2, 2))
plt.legend(loc="best")
plt.title("Grad 1")


ax = plt.subplot(1, 3, 2)
poly = PolynomialFeatures(degree = 5)
pipeline = Pipeline([("polynomial_features", poly), ("linear_regression", linear)])
pipeline.fit(X[:, np.newaxis], y)
X_test = np.linspace(0, 2*math.pi)
plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Vorhersage", color='red')
plt.plot(X_test, np.cos(X_test), label="cos", color='green')
plt.scatter(X, y, edgecolor='b', s=20, label="Datenstze")
plt.xlim((0, 2*math.pi))
plt.ylim((-2, 2))
plt.legend(loc="best")
plt.title("Grad 5")

ax = plt.subplot(1, 3, 3)
poly = PolynomialFeatures(degree = 20)
pipeline = Pipeline([("polynomial_features", poly), ("linear_regression", linear)])
pipeline.fit(X[:, np.newaxis], y)
X_test = np.linspace(0, 2*math.pi)
plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Vorhersage", color='red')
plt.plot(X_test, np.cos(X_test), label="cos", color='green')
plt.scatter(X, y, edgecolor='b', s=20, label="Datenstze")
plt.xlim((0, 2*math.pi))
plt.ylim((-2, 2))
plt.legend(loc="best")
plt.title("Grad 20")


plt.show()
```
